```python
# Import modules
import sys
import os
import re
import copy
from os import listdir
from os.path import isfile, join, isdir

# Setting paths
notebooks_path = "./notebooks/markdown/exp/"
content_path = "./content/"
filename_in = [f for f in listdir(notebooks_path) if isfile(join(notebooks_path, f))][0]

# Import notebook file
with open(notebooks_path + filename_in, 'r', encoding = 'utf8') as file:
    doc = file.read()
    file.close()
```

```python
doc
```

```python
table_regex = re.compile('\n<div>\n<style scoped>.*?</div>\n', re.DOTALL)
```

```python
tables = re.findall(table_regex, doc)
```

```python
table_dict = {}
counter = 1

for table in tables:
    table_placeholder = "{[{TABLE " + str(counter) + "}]}"
    table_dict[table_placeholder] = table
    counter += 1
```

```python
table_dict_rev = {v: k for k, v in table_dict.items()}
```

```python
for key in table_dict_rev.keys():
    doc = doc.replace(key, table_dict_rev[key])
```

```python
with open(notebooks_path + '/python-for-data-analysis_tablefix.md', 'w', encoding = "utf8") as outfile:
    outfile.write(doc)
    outfile.close()
```
